```python
x = np.arange(-100, 100)

def loss(x, m=10, c=1.25, p=2):
    d = c * m**(2-2*p)
    
    losses = np.array(x)
    
    overest = x > 0 
    losses[overest] = np.power(x[overest], 2)
    
    in_margin = np.logical_and(-m < x, x < 0)
    losses[in_margin] = c * np.power(x[in_margin], 2)
    
    below_margin = x < -m
    losses[below_margin] = d * np.power(x[below_margin], 2*p)
    
    return losses

y = loss(x, m=25)
y2 = x**2 * np.power(np.sign(x)-0.95, 2)
plt.plot(x, y)
plt.plot(x, y2, c='red')
```
